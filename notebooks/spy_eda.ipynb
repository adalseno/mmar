```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
spy = pd.read_csv("../data/raw/spy.csv", parse_dates=True, index_col=0)
```

```python
t1 = spy.index[-200]
t2 = spy.index[-350]
t3 = spy.index[-500]
```

```python
spy.loc["2016-01-01":"2023-12-31", "Close"].plot(
    figsize=(12, 6), xlabel="Date", ylabel="Price", title=" SPY prices and test sizes"
)
plt.axvline(t1, ls="--", c="r")
plt.axvline(t2, ls="--", c="g")
plt.axvline(t3, ls="--", c="y")
plt.annotate(
    "Test size = 500",
    (t3, 350),
    ("2021-01-01", 300),
    arrowprops=dict(facecolor="black", headwidth=5, width=1, shrink=0.01),
)
plt.annotate(
    "Test size = 350",
    (t2, 350),
    ("2023-01-01", 200),
    arrowprops=dict(facecolor="black", headwidth=5, width=1, shrink=0.01),
)
plt.annotate(
    "Test size = 200",
    (t1, 350),
    ("2023-06-01", 300),
    arrowprops=dict(facecolor="black", headwidth=5, width=1, shrink=0.01),
)
plt.show()
```

```python
np.log(spy.loc["2016-01-01":"2023-12-31", "Close"]).diff().dropna().plot(
    figsize=(12, 6),
    xlabel="Date",
    ylabel="Log return",
    title=" SPY log returns and test sizes",
)
plt.axvline(t1, ls="--", c="r")
plt.axvline(t2, ls="--", c="g")
plt.axvline(t3, ls="--", c="y")
plt.annotate(
    "Test size = 500",
    (t3, -0.025),
    ("2021-01-01", -0.075),
    arrowprops=dict(facecolor="black", headwidth=5, width=1, shrink=0.01),
)
plt.annotate(
    "Test size = 350",
    (t2, -0.025),
    ("2023-01-01", -0.1),
    arrowprops=dict(facecolor="black", headwidth=5, width=1, shrink=0.01),
)
plt.annotate(
    "Test size = 200",
    (t1, -0.025),
    ("2023-06-01", -0.075),
    arrowprops=dict(facecolor="black", headwidth=5, width=1, shrink=0.01),
)
plt.show()
```
